The sentence_transformers is the entry point for using the Sentence-BERT (SBERT) framework. This library is designed to map sentences, paragraphs, or even entire documents into a dense vector space (a list of numbers called "embeddings").

Unlike older models that just looked at keywords, SBERT captures the semantic meaning of your text. If two sentences mean the same thing but use different words, SBERT will place their vectors very close together in that mathematical space.

**When to Use It**
- You use this when your task requires understanding the intent or meaning of text, rather than just matching exact words. Common scenarios include:

- Semantic Search: Finding documents that answer a question, even if the documents don't contain the exact words from the question.

- Clustering: Grouping similar customer support tickets, news articles, or survey responses automatically.

- Similarity Analysis: Determining how similar two pieces of text are (e.g., checking for plagiarism or duplicate questions).

Recommendation Systems: Suggesting content based on the "flavor" of what a user has read previously.

```python
from sentence_transformers import SentenceTransformer, util

# 1. Load a pre-trained model
# 'all-MiniLM-L6-v2' is a great balance of speed and accuracy
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Define your text data
sentences = [
    "The cat sits outside", 
    "A man is playing guitar", 
    "I love pasta",
    "A feline is resting outdoors"
]

# 3. Encode the sentences into embeddings (numerical representations)
embeddings = model.encode(sentences)

# 4. Compare sentences
# Let's compare the first sentence with the last one (they mean the same thing)
sim_score = util.cos_sim(embeddings[0], embeddings[3])

print(f"Similarity score between '{sentences[0]}' and '{sentences[3]}': {sim_score.item():.4f}")
```

This code uses OpenAI’s CLIP (Contrastive Language–Image Pretraining) model to convert an image into a "numerical fingerprint," known as an embedding [1.1.1, 1.4.1].

This embedding is a high-dimensional vector (a long list of numbers) that captures the semantic meaning of the image [1.1.2, 1.4.3]. Because CLIP was trained on millions of image-text pairs, it understands visual concepts like "cat," "outdoors," or "playing" in a way that allows for tasks like semantic search or clustering [1.1.2, 1.2.1, 1.4.3].

CLIPProcessor: Think of this as the "translator." It takes raw data (like your image) and transforms it into the specific numerical format (pixel values, normalization, resizing) that the model expects [1.3.1].

CLIPModel: This loads the actual neural network architecture [1.3.1]. vit-base-patch32 refers to a specific version of the Vision Transformer architecture [1.3.1].

The processor converts the PIL image object into a PyTorch tensor (return_tensors="pt"). It ensures the image is resized and normalized exactly how the model was trained to see it [1.3.1, 1.4.1].

**Semantic search (Similarity Search)**

Semantic search is the process of retrieving information based on the meaning of the user's query rather than just matching keywords. It works by converting both queries and documents into embeddings (vectors) and then calculating the "distance" between those vectors in a multi-dimensional space.

1. Cosine Similarity
Cosine similarity measures the angle between two vectors. It ignores the magnitude (length) of the vectors and focuses entirely on their orientation.

    - Best for: Most NLP tasks where you care about the direction/topic of the text.

    - Range: -1 (opposite) to 1 (identical).

2. Dot Product
The dot product is the sum of the products of the corresponding entries of two sequences of numbers. If vectors are "normalized" (length = 1), the dot product is exactly the same as the cosine similarity.

    - Best for: Extremely fast retrieval, especially when using vector databases (like Milvus or Pinecone).

    - Range: Can be any number depending on vector magnitude.

3. Euclidean Distance (L2)This measures the "straight-line" distance between two points in space. If the result is 0, the vectors are identical. 
    - Best for: Situations where the magnitude of the vector actually matters (e.g., if a long document's word count changes the meaning).
    - Range: 0 to $\infty$. Lower scores mean more similar.

```python
from sentence_transformers import SentenceTransformer, util
import torch

# Load a lightweight model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Define documents and a query
documents = ["A cat is sitting on a chair", "The car is parked on the street"]
query = "A kitten is on furniture"

# Convert everything to vectors
doc_embeddings = model.encode(documents)
query_embedding = model.encode(query)

# 1. Convert to tensors if they aren't already
query_tensor = torch.tensor(query_embedding)
doc_tensor = torch.tensor(doc_embeddings)

# 2. Add an extra dimension if necessary to make them 2D (1 x N)
if query_tensor.dim() == 1:
    query_tensor = query_tensor.unsqueeze(0)
    
if doc_tensor.dim() == 1:
    doc_tensor = doc_tensor.unsqueeze(0)
```

```python
# Cosine Similarity is the most popular for semantic search
cosine_scores = util.cos_sim(query_embedding, doc_embeddings)
print(f"Cosine Scores: {cosine_scores}")
```

```python
# 3. Perform the matrix multiplication
# We transpose the document tensor so the dimensions align for multiplication
dot_scores = torch.mm(query_tensor, doc_tensor.transpose(0, 1))

print(f"Dot Product Scores: {dot_scores}")
```

```python
# Now cdist will work
l2_distances = torch.cdist(query_tensor, doc_tensor, p=2)

print(f"Euclidean Distances: {l2_distances}")
```

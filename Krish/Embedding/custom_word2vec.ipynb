```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
```

```python
story = []
for file_name in os.listdir(dir_path):
    if file_name.endswith('.txt'):
        fname = os.path.join(dir_path, file_name)
        with open(fname, encoding='unicode_escape') as f:
            corpus = f.read()
        raw_content = sent_tokenize(corpus)
        for sent in raw_content:
            story.append(simple_preprocess(sent))
```

```python
story[0]
```

```python
custom_model = Word2Vec(
    vector_size=10,
    min_count=1,
    window=10
)
```

```python
custom_model.build_vocab(story)
```

```python
custom_model.wv.index_to_key
```

```python
custom_model.corpus_count
```

```python
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=10)
```

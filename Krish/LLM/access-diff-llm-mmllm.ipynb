```python
import time
from langchain_groq import ChatGroq

def get_response(question, max_retries=3, timeout=10):
    retries = 0
    while retries < max_retries:
        try:
            llm = ChatGroq(
                model="meta-llama/llama-4-scout-17b-16e-instruct",
                # Add your Groq API key here if required
                # api_key="YOUR_API_KEY"
            )
            response = llm.invoke(question, timeout=timeout)
            print(response.content)
            return
        except Exception as e:
            print(f"Error occurred: {e}")
            retries += 1
            time.sleep(1)  # wait for 1 second before retrying

    print("Failed to connect after retries.")

# Test the function
question = "What is the capital of India?"
get_response(question)
```

```python
import os
import google.generativeai as genai
from PIL import Image
import io

# --- SECURE CONFIGURATION ---
# 1. Get your free API key from: https://aistudio.google.com/
# 2. Set it as an environment variable named GOOGLE_API_KEY
google_api_key = os.environ.get("GOOGLE_API_KEY")

if not google_api_key:
    print("Error: GOOGLE_API_KEY not found in environment variables.")
    print("Please set it before running the script.")
    exit()

# Configure the SDK with your API key
genai.configure(api_key=google_api_key)

# Initialize the model. Use 'gemini-1.5-pro' for the best image generation quality.
model = genai.GenerativeModel('gemini-1.5-pro')

def generate_image_gemini(prompt, filename="output_gemini.png"):
    print(f"Generating image for prompt: '{prompt}' (this may take a moment...)")
    try:
        # Generate content. Gemini expects a prompt that explicitly asks for image generation.
        response = model.generate_content(
            f"Create a highly detailed image based on this prompt: {prompt}",
            # Important: Tell the model to output an image
            generation_config=genai.types.GenerationConfig(
                # This parameter is what triggers the image generation capability
                # It's specific to the Gemini 1.5 Pro model in AI Studio
                response_mime_type="image/png"
            )
        )
        
        # The response contains the image data
        if response.candidates and response.candidates[0].content.parts[0].inline_data:
             image_bytes = response.candidates[0].content.parts[0].inline_data.data
             
             # Convert bytes to a PIL Image object
             image = Image.open(io.BytesIO(image_bytes))
             
             # Save the image
             image.save(filename)
             print(f"Successfully saved image to: {filename}")
             # image.show()
        else:
             print("No image was generated in the response. This might be due to content safety filters.")

    except Exception as e:
        print(f"An error occurred during Google AI Studio API generation: {e}")

# --- MAIN EXECUTION ---
# Example 1: A simple scene
# generate_image_gemini("A serene mountain landscape with a crystal clear lake reflecting the sky.")

# Example 2: An artistic style
generate_image_gemini("An oil painting of a bowl of fruit in the style of Van Gogh.")
```

```python
from langchain_huggingface import HuggingFaceEndpoint

llm = HuggingFaceEndpoint(
    #repo_id="deepseek-ai/DeepSeek-R1-0528",
    repo_id="deepseek-ai/DeepSeek-V4-Pro",
    task="text-generation",
    max_new_tokens=512
)

from langchain_huggingface import ChatHuggingFace
from langchain_core.messages import SystemMessage, HumanMessage
messages = [
    SystemMessage(content="You're a helpful assistant"),
    HumanMessage(content="What happens when an unstoppable force meets an immovable object?")
]

chat_model = ChatHuggingFace(llm=llm)
ai_msg = chat_model.invoke(messages)
print(ai_msg.content)
```
